<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 4.3 Cyclic Groups of Complex Numbers"><span class="type">Section</span><span class="codenumber">4.3</span><span class="title">Cyclic Groups of Complex Numbers</span></h2></div>

<div class="mathbook-content"><p id="p-622">The <dfn class="terminology">complex numbers</dfn> are defined as</p><div class="displaymath">
\begin{equation*}
{\mathbb C} = \{ a + bi : a, b \in {\mathbb R} \},
\end{equation*}
</div><p>where $i^2 = -1\text{.}$  If $z = a + bi\text{,}$ then $a$ is the <dfn class="terminology">real part</dfn> of $z$ and $b$ is the <dfn class="terminology">imaginary part</dfn> of $z\text{.}$</p></div>

<div class="mathbook-content"><p id="p-623">To add two complex numbers $z=a+bi$ and $w= c+di\text{,}$ we just add the corresponding real and imaginary parts:</p><div class="displaymath">
\begin{equation*}
z + w=(a + bi ) + (c + di)  =  (a + c) + (b + d)i.
\end{equation*}
</div><p>Remembering that $i^2 = -1\text{,}$  we multiply complex numbers just like polynomials. The product of $z$ and $w$ is</p><div class="displaymath">
\begin{equation*}
(a + bi )(c + di)  =   ac + bdi^2 + adi + bci =  (ac -bd) +(ad + bc)i.
\end{equation*}
</div></div>

<div class="mathbook-content"><p id="p-624">Every nonzero complex number $z = a +bi$ has a multiplicative inverse; that is, there exists a $z^{-1} \in {\mathbb C}^\ast$ such that $z z^{-1} = z^{-1} z = 1\text{.}$ If $z = a + bi\text{,}$ then</p><div class="displaymath">
\begin{equation*}
z^{-1} = \frac{a-bi}{ a^2 + b^2  }.
\end{equation*}
</div><p>The <dfn class="terminology">complex conjugate</dfn> of a complex number $z = a + bi$ is defined to be $\overline{z} = a- bi\text{.}$  The <dfn class="terminology">absolute value</dfn> or <dfn class="terminology">modulus</dfn> of  $z = a + bi$  is $|z| = \sqrt{a^2 + b^2}\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-cyclic-complex-add"><h6 class="heading"><span class="type">Example</span><span class="codenumber">4.3.1</span></h6><p id="p-625">Let $z = 2 + 3i$ and $w = 1-2i\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
z + w = (2 + 3i) + (1 - 2i) = 3 + i
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
z  w = (2 + 3i)(1 - 2i ) = 8 - i.
\end{equation*}
</div><p>Also,</p><div class="displaymath">
\begin{align*}
z^{-1} & = \frac{2}{13} - \frac{3}{13}i\\
|z| & = \sqrt{13}\\
\overline{z} & = 2-3i.
\end{align*}
</div></article></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-rectcoord"><img src="images/cyclic-complex-rectangular.svg" width="100%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">4.3.2</span>Rectangular coordinates of a complex number</figcaption></figure></div>

<div class="mathbook-content"><p id="p-626">There are several ways of graphically representing complex numbers. We can represent a complex number $z = a +bi$ as an ordered pair on the $xy$ plane where $a$ is the $x$ (or real) coordinate and $b$ is the $y$ (or imaginary) coordinate. This is called the <dfn class="terminology">rectangular</dfn> or <dfn class="terminology">Cartesian</dfn> representation. The rectangular representations of $z_1 = 2 + 3i\text{,}$ $z_2 = 1 - 2i\text{,}$ and $z_3 = - 3 + 2i$ are depicted in Figure <a href="section-13.ipynb#figure-rectcoord" class="xref" alt="Figure 4.3.2 " title="Figure 4.3.2 ">Figure 4.3.2</a>.</p></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-polarcoord"><img src="images/cyclic-complex-polar.svg" width="100%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">4.3.3</span>Polar coordinates of a complex number</figcaption></figure></div>

<div class="mathbook-content"><p id="p-627">Nonzero complex numbers can also be represented using <dfn class="terminology">polar coordinates</dfn>.  To specify  any nonzero point on the plane, it suffices to give an angle $\theta$ from the positive $x$ axis in the counterclockwise direction and a distance $r$ from the origin, as in  Figure <a href="section-13.ipynb#figure-polarcoord" class="xref" alt="Figure 4.3.3 " title="Figure 4.3.3 ">Figure 4.3.3</a>. We can see that</p><div class="displaymath">
\begin{equation*}
z = a + bi = r( \cos \theta + i \sin \theta).
\end{equation*}
</div><p>Hence,</p><div class="displaymath">
\begin{equation*}
r = |z| = \sqrt{a^2 + b^2}
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{align*}
a & = r \cos \theta\\
b & = r \sin \theta.
\end{align*}
</div><p>We sometimes abbreviate $r( \cos \theta + i \sin \theta)$ as $r \cis \theta\text{.}$  To assure that the representation of $z$ is  well-defined, we also require that $0^{\circ} \leq \theta \lt 360^{\circ}\text{.}$  If the measurement is in radians, then $0 \leq \theta \lt2 \pi\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-cyclic-polar"><h6 class="heading"><span class="type">Example</span><span class="codenumber">4.3.4</span></h6><p id="p-628">Suppose that $z = 2 \cis  60^{\circ}\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
a = 2 \cos 60^{\circ} = 1
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
b = 2 \sin 60^{\circ} = \sqrt{3}.
\end{equation*}
</div><p>Hence, the rectangular representation is $z = 1+\sqrt{3}\, i\text{.}$</p><p id="p-629">Conversely, if we are given a rectangular representation of a complex number, it is often useful to know the number's polar representation. If $z = 3 \sqrt{2} - 3 \sqrt{2}\, i\text{,}$ then</p><div class="displaymath">
\begin{equation*}
r = \sqrt{a^2 + b^2} = \sqrt{36 } = 6
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
\theta = \arctan \left( \frac{b}{a} \right) = \arctan( - 1) = 315^{\circ},
\end{equation*}
</div><p>so $3 \sqrt{2} - 3 \sqrt{2}\, i=6 \cis  315^{\circ}\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-630">The polar representation of a complex number makes it easy to find products and powers of complex numbers.  The proof of the following proposition is straightforward and is left as an exercise.</p></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-proposition-polar-mult" id="proposition-polar-mult"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">4.3.5</span></h6></a></article><div id="hk-proposition-polar-mult" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-631">Let $z = r \cis \theta$ and $w = s \cis \phi$ be two nonzero complex numbers. Then</p><div class="displaymath">
\begin{equation*}
zw = r s \cis( \theta + \phi).
\end{equation*}
</div></article></div></div>

<div class="mathbook-content"><article class="example-like" id="example-cyclic-polar-mult"><h6 class="heading"><span class="type">Example</span><span class="codenumber">4.3.6</span></h6><p id="p-632">If $z =  3 \cis( \pi / 3 )$ and $w = 2 \cis(\pi / 6 )\text{,}$ then $zw = 6 \cis( \pi / 2 ) = 6i\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-theorem-demoivre" id="theorem-demoivre"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">4.3.7</span><span class="title">DeMoivre</span></h6></a></article><div id="hk-theorem-demoivre" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-633">Let $z = r \cis  \theta$ be a nonzero complex number. Then</p><div class="displaymath">
\begin{equation*}
[r \cis \theta  ]^n = r^n \cis( n \theta)
\end{equation*}
</div><p>for $n = 1, 2, \ldots\text{.}$</p></article><article class="proof" id="proof-25"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-634">We will use induction on $n$ (see <a href="section-math-induction.ipynb" class="xref" alt="Section 2.1 Mathematical Induction" title="Section 2.1 Mathematical Induction">Section 2.1</a>). For $n = 1$ the theorem is trivial. Assume that the theorem is true for all $k$ such that $1  \leq k \leq n\text{.}$ Then</p><div class="displaymath">
\begin{align*}
z^{n+1} & = z^n z\\
& = r^n( \cos  n \theta + i \sin n \theta ) r( \cos \theta + i\sin \theta )\\
& = r^{n+1} [( \cos n \theta \cos \theta - \sin n \theta \sin \theta ) + i ( \sin n \theta \cos \theta + \cos n \theta \sin \theta)]\\
& = r^{n+1} [ \cos( n \theta + \theta) + i \sin( n \theta + \theta) ]\\
& = r^{n+1} [ \cos( n +1) \theta + i \sin( n+1) \theta  ].
\end{align*}
</div></article></div></div>

<div class="mathbook-content"><article class="example-like" id="example-50"><h6 class="heading"><span class="type">Example</span><span class="codenumber">4.3.8</span></h6><p id="p-635">Suppose that $z= 1+i$ and we wish to compute $z^{10}\text{.}$ Rather than computing $(1 + i)^{10}$ directly, it is much easier to switch to polar coordinates and calculate $z^{10}$ using DeMoivre's Theorem:</p><div class="displaymath">
\begin{align*}
z^{10} & = (1+i)^{10}\\
& = \left( \sqrt{2} \cis \left( \frac{\pi }{4} \right) \right)^{10}\\
& = ( \sqrt{2}\, )^{10} \cis \left( \frac{5\pi }{2} \right)\\
& = 32  \cis \left( \frac{\pi }{2} \right)\\
& = 32i.
\end{align*}
</div></article></div>

<div class="mathbook-content"><p id="p-636">The multiplicative group of the complex numbers, ${\mathbb C}^*\text{,}$ possesses some interesting subgroups.  Whereas ${\mathbb Q}^*$ and ${\mathbb R}^*$ have no interesting subgroups of finite order, ${\mathbb C}^*$ has many. We first consider the <dfn class="terminology">circle group</dfn>,</p><div class="displaymath">
\begin{equation*}
{\mathbb T} = \{ z \in {\mathbb C} : |z| = 1 \}.
\end{equation*}
</div><p>The following proposition is a direct result of Proposition <a href="section-13.ipynb#proposition-polar-mult" class="xref" alt="Proposition 4.3.5 " title="Proposition 4.3.5 ">Proposition 4.3.5</a>.</p></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-proposition-13" id="proposition-13"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">4.3.9</span></h6></a></article><div id="hk-proposition-13" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-637">The circle group is a subgroup of  ${\mathbb C}^*\text{.}$</p></article></div></div>

<div class="mathbook-content"><p id="p-638">Although the circle group has infinite order, it has many interesting finite subgroups. Suppose that $H = \{ 1, -1, i, -i \}\text{.}$ Then $H$ is a subgroup of the circle group. Also, $1\text{,}$ $-1\text{,}$ $i\text{,}$ and $-i$ are exactly those complex numbers that satisfy the equation $z^4 = 1\text{.}$ The complex numbers satisfying the equation $z^n=1$ are called the <dfn class="terminology">$n$th roots of unity</dfn>.</p></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-theorem-16" id="theorem-16"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">4.3.10</span></h6></a></article><div id="hk-theorem-16" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-639">If $z^n = 1\text{,}$ then the $n$th roots of unity are</p><div class="displaymath">
\begin{equation*}
z = \cis\left( \frac{2 k \pi}{n } \right),
\end{equation*}
</div><p>where $k = 0, 1, \ldots, n-1\text{.}$ Furthermore, the $n$th roots of unity form a cyclic subgroup of ${\mathbb T}$ of order $n$</p></article><article class="proof" id="proof-26"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-640">By DeMoivre's Theorem,</p><div class="displaymath">
\begin{equation*}
z^n = \cis \left( n \frac{2 k \pi}{n } \right) = \cis( 2 k \pi ) = 1.
\end{equation*}
</div><p>The $z$'s are distinct since the numbers $2 k \pi /n$ are all distinct and are greater than or equal to 0 but less than $2 \pi\text{.}$  We will leave the proof that the $n$th roots of unity form a cyclic subgroup of ${\mathbb T}$ as an exercise.</p></article></div></div>

<div class="mathbook-content"><p id="p-641">A generator for the group of the $n$th roots of unity is called a <dfn class="terminology">primitive $n$th root of unity</dfn>.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-roots_unity"><h6 class="heading"><span class="type">Example</span><span class="codenumber">4.3.11</span></h6><p id="p-642">The 8th roots of unity can be represented as eight equally spaced points on the unit circle (Figure <a href="section-13.ipynb#figure-rtsunity" class="xref" alt="Figure 4.3.12 " title="Figure 4.3.12 ">Figure 4.3.12</a>).  The primitive 8th roots of unity are</p><div class="displaymath">
\begin{align*}
\omega & = \frac{\sqrt{2}}{2}  + \frac{\sqrt{2}}{2} i\\
\omega^3 & = -\frac{\sqrt{2}}{2}  + \frac{\sqrt{2}}{2} i\\
\omega^5 & = -\frac{\sqrt{2}}{2}  - \frac{\sqrt{2}}{2} i\\
\omega^7 & = \frac{\sqrt{2}}{2}  - \frac{\sqrt{2}}{2}i. 
\end{align*}
</div></article></div>
