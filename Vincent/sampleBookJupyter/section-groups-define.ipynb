<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 3.2 Definitions and Examples"><span class="type">Section</span><span class="codenumber">3.2</span><span class="title">Definitions and Examples</span></h2><a href="section-groups-define.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-409">The integers mod $n$ and the symmetries of a triangle or a rectangle are examples of groups.  A <dfn class="terminology">binary operation</dfn> or <dfn class="terminology">law of composition</dfn> on a set $G$ is a function $G \times G \rightarrow G$ that assigns to each pair $(a,b) \in G \times G$ a unique element $a \circ b\text{,}$ or $ab$ in $G\text{,}$ called the composition of $a$ and $b\text{.}$  A <dfn class="terminology">group</dfn> $(G, \circ )$ is a set $G$ together with a law of composition $(a,b) \mapsto a \circ b$ that satisfies the following axioms.</p><ul class="disc"><li id="li-89"><p id="p-410">The law of composition is <dfn class="terminology">associative</dfn>. That is,</p><div class="displaymath">
\begin{equation*}
(a \circ b) \circ c = a \circ (b \circ c) 
\end{equation*}
</div><p>for $a, b, c \in G\text{.}$</p></li><li id="li-90"><p id="p-411">There exists an element $e \in G\text{,}$ called the <dfn class="terminology">identity element</dfn>, such that for any element $a \in G$</p><div class="displaymath">
\begin{equation*}
e \circ a = a \circ e = a. 
\end{equation*}
</div></li><li id="li-91"><p id="p-412">For each element $a \in G\text{,}$ there exists an <dfn class="terminology">inverse element</dfn> in G, denoted by $a^{-1}\text{,}$ such that</p><div class="displaymath">
\begin{equation*}
a \circ a^{-1} = a^{-1} \circ a = e. 
\end{equation*}
</div></li></ul></div>

<div class="mathbook-content"><p id="p-413">A group $G$ with the property that $a \circ b = b \circ a$ for all $a, b \in G$ is called <dfn class="terminology">abelian</dfn> or <dfn class="terminology">commutative</dfn>.  Groups not satisfying this property are said to be  <dfn class="terminology">nonabelian</dfn> or <dfn class="terminology">noncommutative</dfn>.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-integers"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.2.1</span></h6><p id="p-414">The integers ${\mathbb Z } = \{ \ldots , -1, 0, 1, 2, \ldots \}$ form a group under the operation of addition.  The binary operation on two integers $m, n \in {\mathbb Z}$ is just their sum.  Since the integers under addition already have a well-established notation, we will use the operator $+$ instead of $\circ\text{;}$ that is, we shall write $m + n$ instead of $m \circ n\text{.}$  The identity is 0, and the inverse of $n \in {\mathbb Z}$ is written as $-n$ instead of $n^{-1}\text{.}$  Notice that the set of integers under addition have the additional property that $m + n = n + m$ and therefore form an abelian group.</p></article></div>

<div class="mathbook-content"><p id="p-415">Most of the time we will write $ab$ instead of $a \circ b\text{;}$ however, if the group already has a natural operation such as addition in the integers, we will use that operation.  That is, if we are adding two integers, we still write $m + n\text{,}$ $-n$ for the inverse, and 0 for the identity as usual. We also write $m - n$ instead of $m + (-n)\text{.}$</p></div>

<div class="mathbook-content"><p id="p-416">It is often convenient to describe a group in terms of an addition or multiplication table.  Such a table is called a <dfn class="terminology">Cayley table</dfn>.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-z5"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.2.2</span></h6><p id="p-417">The integers mod $n$ form a group under addition modulo $n\text{.}$  Consider ${\mathbb Z}_5\text{,}$ consisting of the equivalence classes of the integers 0, 1, 2, 3, and 4.  We define the group operation on ${\mathbb Z}_5$ by modular addition.  We write the binary operation on the group additively; that is, we write $m + n\text{.}$  The element 0 is the identity of the group and each element in ${\mathbb Z}_5$ has an inverse. For instance, $2 + 3 = 3 + 2 = 0\text{.}$  Figure <a href="section-groups-define.ipynb#figure-z5-cayley" class="xref" alt="Figure 3.2.3 " title="Figure 3.2.3 ">Figure 3.2.3</a> is a Cayley table for ${\mathbb Z}_5$  (Contributed by <a href="contributors.ipynb#rbeezer" class="xref" alt="Contributor  " title="Contributor  ">Robert Beezer</a>).  By Proposition <a href="section-mod-n-sym.ipynb#proposition-zn-equiv-classes" class="xref" alt="Proposition 3.1.4 " title="Proposition 3.1.4 ">Proposition 3.1.4</a>, ${\mathbb Z}_n = \{0, 1, \ldots, n-1 \}$ is a group under the binary operation of addition mod $n\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-z6-mult"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.2.4</span></h6><p id="p-419">Not every set with a binary operation is a group.  For example, if we let modular multiplication be the binary operation on ${\mathbb Z}_n\text{,}$ then ${\mathbb Z}_n$ fails to be a group.  The element 1 acts as a group identity since $1 \cdot k = k \cdot 1 = k$ for any $k \in {\mathbb Z}_n\text{;}$ however, a multiplicative inverse for $0$ does not exist since $0 \cdot k = k \cdot 0 = 0$ for every $k$ in ${\mathbb Z}_n\text{.}$  Even if we consider the set ${\mathbb Z}_n \setminus \{0 \}\text{,}$ we still may not have a group.  For instance, let $2 \in {\mathbb Z}_6\text{.}$ Then 2 has no multiplicative inverse since</p><div class="displaymath">
\begin{align*}
0 \cdot 2 & = 0 \qquad 1 \cdot 2 = 2\\
2 \cdot 2 & = 4 \qquad 3 \cdot 2 = 0\\
4 \cdot 2 & = 2 \qquad 5 \cdot 2 = 4.
\end{align*}
</div><p id="p-420">By Proposition <a href="section-mod-n-sym.ipynb#proposition-zn-equiv-classes" class="xref" alt="Proposition 3.1.4 " title="Proposition 3.1.4 ">Proposition 3.1.4</a>, every nonzero $k$ does have an inverse in ${\mathbb Z}_n$ if $k$ is relatively prime to $n\text{.}$  Denote the set of all such nonzero elements in ${\mathbb Z}_n$ by $U(n)\text{.}$   Then $U(n)$ is a group called the <dfn class="terminology">group of units</dfn> of ${\mathbb Z}_n\text{.}$  Figure <a href="section-groups-define.ipynb#figure-cayley-u8" class="xref" alt="Figure 3.2.5 " title="Figure 3.2.5 ">Figure 3.2.5</a> is a Cayley table for the group $U(8)\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-nonabelian"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.2.6</span></h6><p id="p-422">The symmetries of an equilateral triangle described in Section <a href="section-mod-n-sym.ipynb" class="xref" alt="Section 3.1 Integer Equivalence Classes and Symmetries" title="Section 3.1 Integer Equivalence Classes and Symmetries">Section 3.1</a> form a nonabelian group. As we observed, it is not necessarily true that $\alpha \beta = \beta \alpha$ for two symmetries $\alpha$ and $\beta\text{.}$  Using Figure <a href="section-mod-n-sym.ipynb#figure-s3" class="xref" alt="Figure 3.1.7 " title="Figure 3.1.7 ">Figure 3.1.7</a>, which is a Cayley table for this group, we can easily check that the symmetries of an equilateral triangle are indeed a group. We will denote this group by either $S_3$ or $D_3\text{,}$ for reasons that will be explained later.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-gl2"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.2.7</span></h6><p id="p-423">We use  ${\mathbb M}_2 ( {\mathbb R})$ to denote the set of all $2 \times 2$ matrices.  Let $GL_2({\mathbb R})$ be the subset of ${\mathbb M}_2 ( {\mathbb R})$ consisting of invertible matrices; that is, a matrix </p><div class="displaymath">
\begin{equation*}
A =
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
\end{equation*}
</div><p>is in  $GL_2( {\mathbb R})$ if there exists a matrix $A^{-1}$ such that $A A^{-1} = A^{-1} A = I\text{,}$ where $I$ is the $2 \times 2$ identity matrix. For $A$ to have an inverse is equivalent to requiring that the determinant of $A$ be nonzero; that is, $\det A = ad - bc \neq 0\text{.}$  The set of invertible matrices forms a group called the <dfn class="terminology">general linear group</dfn>.   The identity of the group is the identity matrix</p><div class="displaymath">
\begin{equation*}
I =
\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}.
\end{equation*}
</div><p>The inverse of $A \in GL_2( {\mathbb R})$ is</p><div class="displaymath">
\begin{equation*}
A^{-1} =
\frac{1}{ad-bc}
\begin{pmatrix}
d & -b \\
-c & a
\end{pmatrix}.
\end{equation*}
</div><p>The product of two invertible matrices is again invertible. Matrix multiplication is associative, satisfying the other group axiom. For matrices it is not true in general that $AB = BA\text{;}$ hence, $GL_2({\mathbb R})$ is another example of a nonabelian group.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-quaterions"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.2.8</span></h6><p id="p-424">Let</p><div class="displaymath">
\begin{align*}
1 & = 
\begin{pmatrix}
1 & 0\\
0 & 1
\end{pmatrix}
\qquad
I
=
\begin{pmatrix}
0 & 1\\
-1 & 0
\end{pmatrix}\\
J & = 
\begin{pmatrix}
0 & i\\
i & 0
\end{pmatrix}
\qquad
K =
\begin{pmatrix}
i & 0\\
0 & -i
\end{pmatrix},
\end{align*}
</div><p>where $i^2 = -1\text{.}$ Then the relations $I^2 = J^2 = K^2 = -1\text{,}$ $IJ=K\text{,}$ $JK = I\text{,}$ $KI = J\text{,}$ $JI = -K\text{,}$ $KJ = -I\text{,}$ and $IK = -J$ hold. The set $Q_8 = \{\pm 1, \pm I, \pm J, \pm K  \}$ is a group called the <dfn class="terminology">quaternion group</dfn>. Notice that $Q_8$ is noncommutative.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-c-star"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.2.9</span></h6><p id="p-425">Let ${\mathbb C}^\ast$be the set of nonzero complex numbers.  Under the operation of multiplication ${\mathbb C}^\ast$ forms a group.  The identity is 1. If $z = a+bi$ is a nonzero complex number, then</p><div class="displaymath">
\begin{equation*}
z^{-1} = \frac{a -bi}{a^2 +b^2}
\end{equation*}
</div><p>is the inverse of $z\text{.}$  It is easy to see that the remaining group axioms hold.</p></article></div>

<div class="mathbook-content"><p id="p-426">A group is <dfn class="terminology">finite</dfn>, or has <dfn class="terminology">finite order</dfn>, if it contains a finite number of elements; otherwise, the group is said to be <dfn class="terminology">infinite</dfn> or to have <dfn class="terminology">infinite order</dfn>. The <dfn class="terminology">order</dfn> of a finite group is the number of elements that it contains. If $G$ is a group containing $n$ elements, we write $|G| = n\text{.}$  The group ${\mathbb Z}_5$ is a finite group of order 5; the integers ${\mathbb Z}$ form an infinite group under addition, and we sometimes write $|{\mathbb Z}| = \infty\text{.}$</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection 3.2.1 Basic Properties of Groups"><span class="type">Subsection</span><span class="codenumber">3.2.1</span><span class="title">Basic Properties of Groups</span></h3></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-proposition-3" id="proposition-3"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">3.2.10</span></h6></a></article><div id="hk-proposition-3" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-427">The identity element in a group $G$ is unique; that is, there exists only one element $e \in G$ such that $eg = ge = g$ for all $g \in G\text{.}$</p></article><article class="proof" id="proof-14"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-428">Suppose that $e$ and $e'$ are both identities in $G\text{.}$ Then $eg = ge = g$ and $e'g = ge' = g$ for all $g \in G\text{.}$ We need to show that $e = e'\text{.}$ If we think of $e$ as the identity, then $ee' = e'\text{;}$ but if $e'$ is the identity, then $ee' = e\text{.}$ Combining these two equations, we have $e = ee' = e'\text{.}$ (<a href="contributors.ipynb#dfarmer" class="xref" alt="Contributor  " title="Contributor  ">David Farmer</a> helped with this proof.)</p></article></div></div>

<div class="mathbook-content"><p id="p-429">Inverses in a group are also unique. If $g'$ and $g''$ are both inverses of an element $g$ in a group $G\text{,}$ then $gg' = g'g = e$ and $gg'' = g''g = e\text{.}$ We want to show that $g' = g''\text{,}$ but $g' = g'e = g'(gg'') = (g'g)g'' = eg'' = g''\text{.}$ We summarize this fact in the following proposition.</p></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-proposition-4" id="proposition-4"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">3.2.11</span></h6></a></article><div id="hk-proposition-4" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-430">If $g$ is any element in a group $G\text{,}$ then the inverse of $g\text{,}$ denoted by $g^{-1}\text{,}$ is unique.</p></article></div></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-proposition-inverse-of-products" id="proposition-inverse-of-products"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">3.2.12</span></h6></a></article><div id="hk-proposition-inverse-of-products" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-431">Let $G$ be a group. If $a, b \in G\text{,}$ then $(ab)^{-1} = b^{-1}a^{-1}\text{.}$</p></article><article class="proof" id="proof-15"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-432">Let $a, b \in G\text{.}$ Then $abb^{-1}a^{-1} = aea^{-1} = aa^{-1} = e\text{.}$ Similarly, $b^{-1}a^{-1}ab = e\text{.}$ But by the previous proposition, inverses are unique; hence, $(ab)^{-1} = b^{-1}a^{-1}\text{.}$</p></article></div></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-proposition-6" id="proposition-6"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">3.2.13</span></h6></a></article><div id="hk-proposition-6" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-433">Let $G$ be a group.  For any $a \in G\text{,}$ $(a^{-1})^{-1} = a\text{.}$</p></article><article class="proof" id="proof-16"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-434">Observe that $a^{-1} (a^{-1})^{-1} = e\text{.}$ Consequently, multiplying both sides of this equation by $a\text{,}$ we have</p><div class="displaymath">
\begin{equation*}
(a^{-1})^{-1} = e (a^{-1})^{-1} = a a^{-1} (a^{-1})^{-1} = ae = a.
\end{equation*}
</div></article></div></div>

<div class="mathbook-content"><p id="p-435">It makes sense to write equations with group elements and group operations. If $a$ and $b$ are two elements in a group $G\text{,}$ does there exist an element $x \in G$ such that $ax = b\text{?}$ If such an $x$ does exist, is it unique? The following proposition answers both of these questions positively.</p></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-proposition-group-equations" id="proposition-group-equations"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">3.2.14</span></h6></a></article><div id="hk-proposition-group-equations" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-436">Let $G$ be a group and $a$ and $b$ be any two elements in $G\text{.}$ Then the equations $ax = b$ and $xa = b$ have unique solutions in $G\text{.}$</p></article><article class="proof" id="proof-17"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-437">Suppose that $ax = b\text{.}$ We must show that such an $x$ exists. Multiplying both sides of $ax = b$ by $a^{-1}\text{,}$ we have $x = ex = a^{-1}ax = a^{-1}b\text{.}$</p><p id="p-438">To show uniqueness, suppose that $x_1$ and $x_2$ are both solutions of $ax = b\text{;}$ then $ax_1 = b = ax_2\text{.}$ So $x_1 = a^{-1}ax_1 = a^{-1}ax_2 = x_2\text{.}$ The proof for the existence and uniqueness of the solution of $xa = b$ is similar.</p></article></div></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-proposition-8" id="proposition-8"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">3.2.15</span></h6></a></article><div id="hk-proposition-8" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-439">If $G$ is a group and $a, b, c \in G\text{,}$ then $ba = ca$ implies $b = c$ and $ab = ac$ implies $b = c\text{.}$</p></article></div></div>

<div class="mathbook-content"><p id="p-440">This proposition tells us that the <dfn class="terminology">right and  left cancellation laws</dfn> are true in groups.  We leave the proof as an exercise.</p></div>

<div class="mathbook-content"><p id="p-441">We can use exponential notation for groups just as we do in ordinary algebra. If $G$ is a group and $g \in G\text{,}$ then we define $g^0 = e\text{.}$ For $n \in {\mathbb N}\text{,}$ we define</p><div class="displaymath">
\begin{equation*}
g^n = \underbrace{g \cdot g \cdots g}_{n \; \text{times}} 
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
g^{-n} = \underbrace{g^{-1} \cdot g^{-1} \cdots g^{-1}}_{n \; \text{times}}.
\end{equation*}
</div></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-theorem-exponent-laws" id="theorem-exponent-laws"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">3.2.16</span></h6></a></article><div id="hk-theorem-exponent-laws" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-442">In a group, the usual laws of exponents hold; that is, for all $g, h \in G\text{,}$</p><ol class="decimal"><li id="li-92"><p id="p-443">$g^mg^n = g^{m+n}$ for all $m, n \in {\mathbb Z}\text{;}$</p></li><li id="li-93"><p id="p-444">$(g^m)^n = g^{mn}$ for all $m, n \in {\mathbb Z}\text{;}$</p></li><li id="li-94"><p id="p-445">$(gh)^n = (h^{-1}g^{-1})^{-n}$ for all $n \in {\mathbb Z}\text{.}$ Furthermore, if $G$ is abelian, then $(gh)^n = g^nh^n\text{.}$</p></li></ol></article></div></div>

<div class="mathbook-content"><p id="p-446">We will leave the proof of this theorem as an exercise. Notice that $(gh)^n \neq g^nh^n$ in general, since the group may not be abelian. If the group is ${\mathbb Z}$ or ${\mathbb Z}_n\text{,}$ we write the group operation additively and the exponential operation multiplicatively; that is, we write $ng$ instead of $g^n\text{.}$ The laws of exponents now become</p><ol class="decimal"><li id="li-95"><p id="p-447">$mg + ng = (m+n)g$ for all $m, n \in {\mathbb Z}\text{;}$</p></li><li id="li-96"><p id="p-448">$m(ng) = (mn)g$ for all $m, n \in {\mathbb Z}\text{;}$</p></li><li id="li-97"><p id="p-449">$m(g + h) = mg + mh$ for all $n \in {\mathbb Z}\text{.}$</p></li></ol></div>

<div class="mathbook-content"><p id="p-450">It is important to realize that the last statement can be made only because ${\mathbb Z}$ and  ${\mathbb Z}_n$ are commutative groups.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection 3.2.2 Historical Note"><span class="type">Subsection</span><span class="codenumber">3.2.2</span><span class="title">Historical Note</span></h3></div>

<div class="mathbook-content"><p id="p-451">Although the first clear axiomatic definition of a group was not given until the late 1800s, group-theoretic methods had been employed before this time in the development of  many areas of mathematics, including geometry and the theory of algebraic equations.</p></div>

<div class="mathbook-content"><p id="p-452">Joseph-Louis Lagrange used group-theoretic methods in a 1770–1771 memoir to study methods of solving polynomial equations.  Later, Évariste Galois (1811–1832) succeeded in developing the mathematics necessary to determine exactly which polynomial equations could be  solved in terms of the polynomials'coefficients. Galois' primary tool was group theory.</p></div>

<div class="mathbook-content"><p id="p-453">The study of geometry was revolutionized in 1872 when Felix Klein proposed that geometric spaces should be studied by examining those properties that are invariant under a transformation of the space. Sophus Lie, a contemporary of Klein, used group theory to study solutions of partial differential equations.  One of the first modern treatments of group theory appeared in William Burnside's <span class="booktitle">The Theory of Groups of Finite Order</span> [1], first published in 1897.</p></div>
