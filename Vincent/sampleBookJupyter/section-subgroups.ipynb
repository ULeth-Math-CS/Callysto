<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection 3.3.1 Definitions and Examples"><span class="type">Subsection</span><span class="codenumber">3.3.1</span><span class="title">Definitions and Examples</span></h3></div>

<div class="mathbook-content"><p id="p-454">Sometimes we wish to investigate smaller groups sitting inside a larger group.   The set of even integers $2{\mathbb Z} = \{\ldots, -2, 0, 2, 4, \ldots \}$ is a group  under the operation of addition. This smaller group sits naturally inside of the group of integers under addition.  We define a <dfn class="terminology">subgroup</dfn> $H$ of a group $G$ to be a subset $H$ of $G$ such that when the group operation of $G$ is restricted to $H\text{,}$ $H$ is a group in its own right. Observe that every group $G$ with at least two elements will always have at least two subgroups, the subgroup consisting of the identity element alone and the entire group itself. The subgroup $H = \{ e \}$ of a group $G$ is called the <dfn class="terminology">trivial subgroup</dfn>.  A subgroup that is a proper subset of $G$ is called a <dfn class="terminology">proper subgroup</dfn>. In many of the examples that we have investigated up to this point, there exist other subgroups besides the trivial and improper subgroups.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-multiplicative-subgroup"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.3.1</span></h6><p id="p-455">Consider the set of nonzero real numbers, ${\mathbb R}^*\text{,}$ with the group operation of multiplication.  The identity of this group is 1 and the inverse of any element $a \in {\mathbb R}^*$ is just $1/a\text{.}$ We will show that</p><div class="displaymath">
\begin{equation*}
{\mathbb Q}^* = \{ p/q : p \, {\rm and }\, q\, {\rm are\, nonzero\, integers} \}
\end{equation*}
</div><p>is a subgroup of ${\mathbb R}^*\text{.}$  The identity of ${\mathbb R}^*$ is 1; however,  $1 = 1/1$ is the quotient of two nonzero integers. Hence, the identity of ${\mathbb R}^*$ is in ${\mathbb Q}^*\text{.}$ Given two elements in ${\mathbb Q}^*\text{,}$ say $p/q$ and $r/s\text{,}$ their product $pr/qs$ is also in ${\mathbb Q}^*\text{.}$ The inverse of any element $p/q \in {\mathbb Q}^*$ is again in ${\mathbb Q}^*$ since $(p/q)^{-1} = q/p\text{.}$ Since multiplication in ${\mathbb R}^*$ is associative, multiplication in ${\mathbb Q}^*$ is associative.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-subgroup-h"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.3.2</span></h6><p id="p-456">Recall that ${\mathbb C}^{\ast}$  is the multiplicative group of nonzero complex numbers. Let $H = \{ 1, -1, i, -i \}\text{.}$ Then $H$ is a subgroup of ${\mathbb C}^{\ast}\text{.}$ It is quite easy to verify  that $H$ is a group under multiplication and that $H \subset {\mathbb C}^{\ast}\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-sl2"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.3.3</span></h6><p id="p-457">Let $SL_2( {\mathbb R})$ be the subset of $GL_2( {\mathbb R })$consisting of matrices of determinant one; that is, a matrix </p><div class="displaymath">
\begin{equation*}
A =
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
\end{equation*}
</div><p>is in $SL_2( {\mathbb R})$ exactly when $ad - bc = 1\text{.}$ To show that $SL_2( {\mathbb R})$ is a subgroup of the general linear group, we must show that it is a group under matrix multiplication.  The $2 \times 2$ identity matrix is in $SL_2( {\mathbb R})\text{,}$ as is the inverse of the matrix $A\text{:}$</p><div class="displaymath">
\begin{equation*}
A^{-1} =
\begin{pmatrix}
d  & -b \\
-c &  a
\end{pmatrix}.
\end{equation*}
</div><p>It remains to show that multiplication is closed; that is, that the product of two matrices of determinant one also has determinant one. We will leave this task as an exercise. The group $SL_2({\mathbb R})$ is called the <dfn class="terminology">special linear group</dfn>.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-gl2-subgroup"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.3.4</span></h6><p id="p-458">It is important to realize that a subset $H$ of a group $G$ can be a group without being a subgroup of $G\text{.}$ For $H$ to be a subgroup of $G$ it must inherit $G$'s binary operation.  The set of all $2 \times 2$ matrices, ${\mathbb M}_2(\mathbb R)\text{,}$ forms a group under the operation of addition. The $2 \times 2$ general linear group is a subset of ${\mathbb M}_2(\mathbb R)$ and is a group under matrix multiplication, but it is not a subgroup of ${\mathbb M}_2(\mathbb R)\text{.}$  If we add two invertible matrices, we do not necessarily obtain another invertible matrix. Observe that</p><div class="displaymath">
\begin{equation*}
\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
+
\begin{pmatrix}
-1 &  0 \\
0  & -1
\end{pmatrix}
=
\begin{pmatrix}
0 & 0 \\
0 & 0
\end{pmatrix}\text{,}
\end{equation*}
</div><p>but the zero matrix is not in $GL_2( {\mathbb R })\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-groups-z2xz2"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.3.5</span></h6><p id="p-459">One way of telling whether or not two groups are the same is by examining their subgroups.  Other than the trivial subgroup and the group itself, the group ${\mathbb Z}_4$ has a single subgroup consisting of the elements 0 and 2. From the group ${\mathbb Z}_2\text{,}$ we can form another group of four elements as follows.  As a set this group is ${\mathbb Z}_2 \times {\mathbb Z}_2\text{.}$ We perform the group operation coordinatewise; that is, $(a,b) + (c,d) = (a+c, b+d)\text{.}$ Figure <a href="section-subgroups.ipynb#figure-z2xz2" class="xref" alt="Figure 3.3.6 " title="Figure 3.3.6 ">Figure 3.3.6</a> is an addition table for ${\mathbb Z}_2 \times {\mathbb Z}_2\text{.}$ Since there are three nontrivial proper subgroups of ${\mathbb Z}_2 \times {\mathbb Z}_2\text{,}$ $H_1 = \{ (0,0), (0,1) \}\text{,}$ $H_2 = \{ (0,0), (1,0) \}\text{,}$ and $H_3 = \{ (0,0), (1,1) \}\text{,}$ ${\mathbb Z}_4$ and ${\mathbb Z}_2 \times {\mathbb Z}_2$ must be different groups.</p></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection 3.3.2 Some Subgroup Theorems"><span class="type">Subsection</span><span class="codenumber">3.3.2</span><span class="title">Some Subgroup Theorems</span></h3></div>

<div class="mathbook-content"><p id="p-461">Let us examine some criteria for determining exactly when a subset of a group is a subgroup.</p></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-proposition-9" id="proposition-9"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">3.3.7</span></h6></a></article><div id="hk-proposition-9" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-462">A subset $H$ of $G$ is a subgroup if and only if it satisfies the following conditions.</p><ol class="decimal"><li id="li-98"><p id="p-463">The identity $e$ of $G$ is in $H\text{.}$</p></li><li id="li-99"><p id="p-464">If $h_1, h_2 \in H\text{,}$ then $h_1h_2 \in H\text{.}$</p></li><li id="li-100"><p id="p-465">If $h \in H\text{,}$ then $h^{-1} \in H\text{.}$</p></li></ol></article><article class="proof" id="proof-18"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-466">First suppose that $H$ is a subgroup of $G\text{.}$  We must show that the three conditions hold.  Since $H$ is a group, it must have an identity $e_H\text{.}$  We must show that $e_H = e\text{,}$ where $e$ is the identity of $G\text{.}$ We know that $e_H e_H = e_H$ and that $ee_H = e_H e = e_H\text{;}$ hence, $ee_H = e_H e_H\text{.}$  By right-hand cancellation, $e =e_H\text{.}$  The second condition holds since a subgroup $H$ is a group.  To prove the third condition, let $h \in H\text{.}$  Since $H$ is a group, there is  an element $h' \in H$ such that $hh' = h'h = e\text{.}$  By the uniqueness  of the inverse in $G\text{,}$ $h' = h^{-1}\text{.}$</p><p id="p-467">Conversely, if the three conditions hold, we must show that $H$ is a group under the same operation as $G\text{;}$ however, these conditions plus the associativity of the binary operation are exactly the axioms stated in the definition of a group.</p></article></div></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-proposition-subgroup" id="proposition-subgroup"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">3.3.8</span></h6></a></article><div id="hk-proposition-subgroup" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-468">Let $H$ be a subset of a group $G\text{.}$  Then $H$ is a subgroup of $G$ if and only if $H \neq \emptyset\text{,}$ and whenever $g, h \in H$ then $gh^{-1}$ is in $H\text{.}$</p></article><article class="proof" id="proof-19"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-469">First assume that $H$ is a subgroup of $G\text{.}$  We wish to show that $gh^{-1} \in H$ whenever $g$ and $h$ are in $H\text{.}$  Since $h$ is in $H\text{,}$ its inverse $h^{-1}$ must also be in $H\text{.}$  Because of the closure of the group operation, $gh^{-1} \in H\text{.}$</p><p id="p-470">Conversely, suppose that $H \subset G$ such that $H \neq \emptyset$ and  $g h^{-1} \in H$ whenever $g, h \in H\text{.}$  If $g \in H\text{,}$ then $gg^{-1} = e$ is in $H\text{.}$  If $g \in H\text{,}$ then $eg^{-1} = g^{-1}$ is also in $H\text{.}$ Now let $h_1, h_2 \in H\text{.}$ We must show that their product is also in $H\text{.}$  However, $h_1(h_2^{-1})^{-1} = h_1 h_2 \in H\text{.}$  Hence, $H$ is a subgroup of $G\text{.}$</p></article></div></div>
