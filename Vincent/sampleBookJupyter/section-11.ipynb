<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 4.1 Cyclic groups"><span class="type">Section</span><span class="codenumber">4.1</span><span class="title">Cyclic groups</span></h2></div>

<div class="mathbook-content"><p id="p-597">Often a subgroup will depend entirely on a single element of the group; that is, knowing that particular element will allow us to compute any other element in the subgroup.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-cyclic-z3"><h6 class="heading"><span class="type">Example</span><span class="codenumber">4.1.1</span></h6><p id="p-598">Suppose that we consider $3 \in {\mathbb Z}$ and look at all multiples (both positive and negative) of 3.  As a set, this is</p><div class="displaymath">
\begin{equation*}
3 {\mathbb Z} = \{ \ldots, -3, 0, 3, 6, \ldots \}.
\end{equation*}
</div><p>It is easy to see that $3 {\mathbb Z}$ is a subgroup of the integers.  This subgroup is completely determined by the element 3 since we can obtain all of the other elements of the group by taking multiples of 3.  Every element in the subgroup is “generated” by 3.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-cyclic-2n"><h6 class="heading"><span class="type">Example</span><span class="codenumber">4.1.2</span></h6><p id="p-599">If $H = \{ 2^n : n \in {\mathbb Z} \}\text{,}$ then $H$ is a subgroup of the multiplicative group of nonzero rational numbers, ${\mathbb Q}^*\text{.}$  If $a = 2^m$ and $b = 2^n$ are in $H\text{,}$ then $ab^{-1} = 2^m 2^{-n} = 2^{m-n}$ is also in $H\text{.}$  By Proposition <a href="section-subgroups.ipynb#proposition-subgroup" class="xref" alt="Proposition 3.3.8 " title="Proposition 3.3.8 ">Proposition 3.3.8</a>, $H$ is a subgroup of ${\mathbb Q}^*$ determined by the element 2.</p></article></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-theorem-cyclic-subgroup" id="theorem-cyclic-subgroup"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">4.1.3</span></h6></a></article><div id="hk-theorem-cyclic-subgroup" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-600">Let $G$ be a group and $a$ be any element in $G\text{.}$  Then the set</p><div class="displaymath">
\begin{equation*}
\langle a \rangle  = \{ a^k : k \in {\mathbb Z} \}
\end{equation*}
</div><p>is a subgroup of $G\text{.}$  Furthermore, $\langle a \rangle$ is the smallest subgroup of $G$ that contains~$a\text{.}$</p></article><article class="proof" id="proof-20"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-601">The identity is in $\langle a \rangle $ since $a^0 = e\text{.}$ If $g$ and $h$ are any two elements in $\langle a \rangle \text{,}$ then by the definition of $\langle a \rangle$ we can write $g = a^m$ and $h = a^n$ for some integers $m$ and $n\text{.}$ So $gh = a^m a^n = a^{m+n}$ is again in $\langle a \rangle \text{.}$ Finally, if $g = a^n$ in $\langle a \rangle \text{,}$ then the inverse $g^{-1} = a^{-n}$ is also in $\langle a \rangle \text{.}$ Clearly, any subgroup $H$ of $G$ containing $a$ must contain all the powers of $a$ by closure; hence, $H$ contains $\langle a \rangle \text{.}$ Therefore, $\langle a \rangle $ is the smallest subgroup of $G$ containing $a\text{.}$</p></article></div></div>

<div class="mathbook-content"><article class="remark-like" id="remark-3"><h6 class="heading"><span class="type">Remark</span><span class="codenumber">4.1.4</span></h6><p id="p-602">If we are using the “+” notation, as in the case of the integers under addition, we write $\langle a \rangle  = \{ na : n \in {\mathbb Z} \}\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-603">For $a \in G\text{,}$ we call $\langle a \rangle $ the <dfn class="terminology">cyclic subgroup</dfn> generated by $a\text{.}$ If $G$ contains some element $a$ such that $G = \langle a \rangle \text{,}$ then $G$ is a <dfn class="terminology">cyclic group</dfn>. In this case $a$ is a <dfn class="terminology">generator</dfn> of $G\text{.}$  If $a$ is an element of a group $G\text{,}$ we define the <dfn class="terminology">order</dfn> of $a$ to be the smallest positive integer $n$ such that $a^n= e\text{,}$ and we write $|a| = n\text{.}$  If there is no such integer $n\text{,}$ we say that the order of $a$ is infinite and  write $|a| = \infty$ to denote the order of $a\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-cyclic-z6"><h6 class="heading"><span class="type">Example</span><span class="codenumber">4.1.5</span></h6><p id="p-604">Notice that a cyclic group can have more than a single generator. Both 1 and 5 generate ${\mathbb Z}_6\text{;}$ hence, ${\mathbb Z}_6$ is a cyclic group. Not every element in a cyclic group is necessarily a generator of the group. The order of $2 \in {\mathbb Z}_6$ is 3. The cyclic subgroup generated by 2 is $\langle 2 \rangle  = \{ 0, 2, 4 \}\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-605">The groups ${\mathbb Z}$ and ${\mathbb Z}_n$ are cyclic groups. The elements 1 and $-1$ are generators for ${\mathbb Z}\text{.}$  We can certainly generate ${\mathbb Z}_n$ with 1 although there may be other generators of ${\mathbb Z}_n\text{,}$ as in the case of ${\mathbb Z}_6\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-cyclic-u9"><h6 class="heading"><span class="type">Example</span><span class="codenumber">4.1.6</span></h6><p id="p-606">The group of units, $U(9)\text{,}$ in ${\mathbb Z}_9$ is a cyclic group.  As a set, $U(9)$ is $\{ 1, 2, 4, 5, 7, 8  \}\text{.}$ The element 2 is a generator for $U(9)$ since</p><div class="displaymath">
\begin{align*}
2^1 & = 2 \qquad 2^2 = 4\\
2^3 & = 8 \qquad 2^4 = 7\\
2^5 & =  5 \qquad 2^6 = 1.
\end{align*}
</div></article></div>

<div class="mathbook-content"><article class="example-like" id="example-cyclic-s3-not-cyclic"><h6 class="heading"><span class="type">Example</span><span class="codenumber">4.1.7</span></h6><p id="p-607">Not every group is a cyclic group.  Consider the symmetry group of an equilateral triangle $S_3\text{.}$  The subgroups of $S_3$ are shown in Figure <a href="section-11.ipynb#figure-subgrps-s3" class="xref" alt="Figure 4.1.8 " title="Figure 4.1.8 ">Figure 4.1.8</a>.  Notice that every subgroup is cyclic; however, no single element generates the entire group.</p></article></div>

<div class="mathbook-content"><article class="theorem-like"><a knowl="" class="id-ref" refid="hk-theorem-cyclic-abelian" id="theorem-cyclic-abelian"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">4.1.9</span></h6></a></article><div id="hk-theorem-cyclic-abelian" class="hidden-content tex2jax_ignore"><article class="theorem-like"><p id="p-608">Every cyclic group is abelian.</p></article><article class="proof" id="proof-21"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-609">Let $G$ be a cyclic group and $a \in G$ be a generator for $G\text{.}$ If $g$ and $h$ are in $G\text{,}$ then they can be written as powers of $a\text{,}$ say $g = a^r$ and $h = a^s\text{.}$ Since</p><div class="displaymath">
\begin{equation*}
g  h = a^r a^s = a^{r+s} = a^{s+r} = a^s a^r = h g,
\end{equation*}
</div><p>$G$ is abelian.</p></article></div></div>
