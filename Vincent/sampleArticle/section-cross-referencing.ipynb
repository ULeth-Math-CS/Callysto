<div class="mathbook-content"><p id="p-557">Cross-references are easy, since that is a key reason for having a highly structured document.  Here is a useful feature if you elect to use it.  Any <code class="code-inline tex2jax_ignore"><xref></code> will “know” what it points to, so you can let it provide the “naming” part of the cross-reference text.  You can turn this on globally with the command-line parameter <code class="code-inline tex2jax_ignore">autoname</code> set to <code class="code-inline tex2jax_ignore">'yes'</code>.  If you do that, you will see most of the names in this document doubled, since the names are written into the source already in most places outside of this section.  Try it and see: use <code class="code-inline tex2jax_ignore">--stringparam autoname 'yes'</code> as an argument to <code class="code-inline tex2jax_ignore">xsltproc</code>.</p></div>

<div class="mathbook-content"><p id="p-558">Moreover, the names themselves will change with the use of the one language dependent file.  And another bonus is that with an autoname, you automatically get a <em class="emphasis">non-breaking</em> space between the name and the reference.  The autoname switch makes no sense for “provisional” cross-references, since there is no information about what they point to.</p></div>

<div class="mathbook-content"><p id="p-559">Here is a reference that has no indication of its type in the source:  <a href="section-fundamental-theorem.ipynb#theorem-FTC" class="xref" alt="Theorem 2.1 The Fundamental Theorem of Calculus" title="Theorem 2.1 The Fundamental Theorem of Calculus">2.1</a>.  So by default you will just see a number that you can click on.  If you use the <code class="code-inline tex2jax_ignore">text="type-global"</code> switch then you should see “Theorem” prepended.  Note that if you changed the theorem to a lemma, then that change would be reflected here automatically when autonaming is in effect.</p></div>

<div class="mathbook-content"><p id="p-560">If you set the autonaming behavior globally, or accept the default behavior, there will still be instances where you want to override that choice.  Simple: just say <code class="code-inline tex2jax_ignore">text="type-global"</code> or <code class="code-inline tex2jax_ignore">text="global"</code> as part of the <code class="code-inline tex2jax_ignore">xref</code>.  Each example below should look the same each time this article is processed, no matter how the global <code class="code-inline tex2jax_ignore">autoname</code> is set.</p><ul class="disc"><li id="li-183"><p id="p-561">No name ever: <a href="interesting-corollary.ipynb#corollary-FTC-derivative" class="xref" alt="Corollary 4.1 " title="Corollary 4.1 ">4.1</a></p></li><li id="li-184"><p id="p-562">Always named: <a href="interesting-corollary.ipynb#corollary-FTC-derivative" class="xref" alt="Corollary 4.1 " title="Corollary 4.1 ">Corollary 4.1</a></p></li></ul></div>

<div class="mathbook-content"><p id="p-563">You might also wish to provide a prefix to a cross-reference and have it incorporated into the text of what you would click on in an electronic version.  So if you make an <code class="code-inline tex2jax_ignore">xref</code> with some content, then that content will prefix the cross-reference within the clickable/pokeable text and be attached with a non-breaking space.  This <code class="code-inline tex2jax_ignore">xref</code> content totally overrides any prefix that might happen otherwise.  So the name of an item (e.g. “corollary”) could be replaced, and if you make a cross-reference with a title as the clickable, then that text can be replaced also.  An example:</p><ul class="disc"><li id="li-185"><p id="p-564">A grand result: <a href="interesting-corollary.ipynb#corollary-FTC-derivative" class="xref" alt="Corollary 4.1 " title="Corollary 4.1 ">Major Corollary 4.1</a></p></li><li id="li-186"><p id="p-565">A grand result: <a href="interesting-corollary.ipynb#corollary-FTC-derivative" class="xref" alt="Corollary 4.1 " title="Corollary 4.1 ">a nice corollary</a></p></li></ul></div>

<div class="mathbook-content"><p id="p-566">Suppose you want to reference two theorems, so you might want to say something like “Theorems 4.6 and 5.2.”  With global autonaming on, you can override the first <code class="code-inline tex2jax_ignore">Theorem</code> by providing the content <code class="code-inline tex2jax_ignore">Theorems</code> on the first <code class="code-inline tex2jax_ignore">xref</code> and <code class="code-inline tex2jax_ignore">text="global"</code> on the second <code class="code-inline tex2jax_ignore">xref</code>.  (With global autonaming off, you will also get what you want/expect.)   Here is the test, which should look correct no matter what the global switch is:  <a href="section-cross-referencing.ipynb" class="xref" alt="Section 17 Cross-Referencing" title="Section 17 Cross-Referencing">Sections 17</a> and <a href="section-internationalization.ipynb" class="xref" alt="Section 18 Internationalization" title="Section 18 Internationalization">18</a>.
