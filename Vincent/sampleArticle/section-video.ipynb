<div class="mathbook-content"><p id="p-543">Embedded videos can make sense for a web version of your document. This is a video promoting the University of Puget Sound to potential new students, in WebM format.  Support is limited to HTML5-capable browsers.  The file format can be MP4, Ogg, or WebM, though this may vary depending upon the browser.  Use a <code class="code-inline tex2jax_ignore"><video></code> element, within either a <code class="code-inline tex2jax_ignore"><figure></code> (numbered, captioned) or a <code class="code-inline tex2jax_ignore"><sidebyside></code> for more precise control.   The <code class="code-inline tex2jax_ignore">@source</code> attribute in this first example <em class="emphasis">does not</em> include an extension, and so three possibilities above will be searched for preferentially (you need only provide the video in one format, but providing additional versions will increase the chances every browser will find a compatible format).</p></div>

<div class="mathbook-content"><p id="p-544">You can replace the “preview” image of a video with one of your own making.  <abbr class="initialism">HTML</abbr> refers to this as the <dfn class="terminology">poster</dfn>, presumably because it advertises the video.  The image you make should be of the same quality as the video, and with the same aspect-ratio, and is presumably one of the frames of the video.  So a screenshot is the likely avenue.  (Maybe we will have advice on how to do this easily, or see <a class="url" href="https://github.com/rbeezer/mathbook/issues/853" target="_blank"><code class="code-inline tex2jax_ignore">https://github.com/rbeezer/mathbook/issues/853</code></a>.)  On the <code class="code-inline tex2jax_ignore"><video></code> tag, include a <code class="code-inline tex2jax_ignore">@preview</code> attribute which is the name of an image file, including a relative path.  (JPEG or PNG formats are best.  JPEG may be smaller for video screenshots, while PNG is lossless and so may work better for diagrams.)</p></div>
