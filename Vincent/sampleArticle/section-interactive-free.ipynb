<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection 15.1 GeoGebra (Cloud)"><span class="type">Subsection</span><span class="codenumber">15.1</span><span class="title">GeoGebra (Cloud)</span></h3></div>

<div class="mathbook-content"><p id="p-525">A Geogebra <dfn class="terminology">material</dfn> is something you might use in a class.  It could also be called an interactive demonstration.  Go browsing at <a class="url" href="https://www.geogebra.org/materials/" target="_blank">www.geogebra.org/materials</a> and find something appropriate for your project.  Or make an account and create your own.</p></div>

<div class="mathbook-content"><p id="p-526">Once you find a material that looks instructive, it will be at a <abbr class="initialism">URL</abbr> such as</p><pre class="code-block tex2jax_ignore">https://www.geogebra.org/m/KGn2d4Qd</pre><p>and you want to pick off the identifier on the end, in this case <code class="code-inline tex2jax_ignore">KGn2d4Qd</code>.  Then author</p><pre class="code-block tex2jax_ignore"><interactive geogebra="KGn2d4Qd" /></pre><p>At this writing (2018-02-04) you will want to place this inside a <code class="code-inline tex2jax_ignore">figure</code>, with a caption, as we do right now with material <code class="code-inline tex2jax_ignore">KGn2d4Qd</code>.</p></div>

<div class="mathbook-content"><p id="p-527">The shape of the material will be fixed, so guess (or measure with an on-screen ruler), the aspect ratio and use that in the <code class="code-inline tex2jax_ignore"><interactive></code> element.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection 15.4 Wolfram CDF (Cloud)"><span class="type">Subsection</span><span class="codenumber">15.4</span><span class="title">Wolfram CDF (Cloud)</span></h3></div>

<div class="mathbook-content"><p id="p-537">You can embed interactive demonstrations created in Wolfram's <dfn class="terminology">Computable Document Format</dfn> so that they will be played with the Wolfram CDF Player™.  Once you create and save a demonstration, you want to determine the UUID that is the identifier of your demonstration.  For example, <a href="section-interactive-free.ipynb#wolfram-cdf-cellular-automata" class="xref" alt="Figure 15.7 " title="Figure 15.7 ">Figure 15.7</a> is identified by <code class="code-inline tex2jax_ignore">9fa2acff-c809-4b7f-a73b-c59ace36affc</code>.  This identifier is enough to create the PreTeXt to embed the demonstration.  See <a class="url" href="https://reference.wolfram.com/language/howto/DeployInteractiveContentInTheWolframCloud.html" target="_blank"><code class="code-inline tex2jax_ignore">https://reference.wolfram.com/language/howto/DeployInteractiveContentInTheWolframCloud.html</code></a> for information about creating your demonstration.</p></div>

<div class="mathbook-content"><p id="p-538"><a class="url" href="http://www.wolfram.com/cdf/adopting-cdf/deploying-cdf/web-delivery-cloud.html" target="_blank"><code class="code-inline tex2jax_ignore">http://www.wolfram.com/cdf/adopting-cdf/deploying-cdf/web-delivery-cloud.html</code></a> explains hosting CDF files at the Wolfram Cloud, and is the source of <a href="section-interactive-free.ipynb#wolfram-cdf-cellular-automata" class="xref" alt="Figure 15.7 " title="Figure 15.7 ">Figure 15.7</a>.  You can learn about powering your CDF with Wolfram Cloud Credits at <a class="url" href="https://www.wolfram.com/cloud-credits/" target="_blank"><code class="code-inline tex2jax_ignore">https://www.wolfram.com/cloud-credits/</code></a>.  CDF is a public format, and the FreeCDF™ license is a variant of a Creative Commons BY-SA license, see <a class="url" href="http://www.wolfram.com/cdf/adopting-cdf/licensing-options.html" target="_blank"><code class="code-inline tex2jax_ignore">http://www.wolfram.com/cdf/adopting-cdf/licensing-options.html</code></a>.</p></div>
