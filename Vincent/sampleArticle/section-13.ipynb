<div class="mathbook-content"><p id="p-453">The next table describes how to construct tables via the <code class="code-inline tex2jax_ignore">tabular</code> element.  The <code class="code-inline tex2jax_ignore">table</code> element may be used to enclose the raw table, so as to associate a caption and get vertical separation with horizontal centering.</p></div>

<div class="mathbook-content"><p id="p-454">The <code class="code-inline tex2jax_ignore">tabular</code> element contains a sequence of <code class="code-inline tex2jax_ignore">row</code> elements, and must contain at least one.  Each <code class="code-inline tex2jax_ignore">row</code> contains a sequence of <code class="code-inline tex2jax_ignore">cell</code> elements and must have the same number in each row (acccounting for the use of the <code class="code-inline tex2jax_ignore">colspan</code> attribute).  The contents of the <code class="code-inline tex2jax_ignore">cell</code> elements are the text to appear in entries of the table.</p></div>

<div class="mathbook-content"><p id="p-455">A sequence of <code class="code-inline tex2jax_ignore">col</code> elements may optionally be used.  But if one appears, then there must be the right number for the width of the table.  They are empty elements always, and just carry information about their respective column.</p></div>

<div class="mathbook-content"><p id="p-456">Where the body of the table below has an entry, it means the attribute may be used on the element, and affects the range of the tabular described by the element.  Employment of an attribute on elements to the right in the table will supersede use on elements to the left.  Generally, every cell has right and bottom borders, but only cells at the left side of the table have a left border and only cells across the top have a top border.  Only one cell has four borders.</p></div>

<div class="mathbook-content"><h6 class="heading"><span class="title">Bully Pulpit: Vertical Rules in Tables</span></h6></div>

<div class="mathbook-content"><p id="p-457">One of the goals of PreTeXt is to gently guide authors towards good choices in the design of their documents, even if we do not claim to know it all ourselves.  Take a close look at that table about tables.  What's missing?  No vertical rules.  Try living without them, you will not really miss them.  If you think you need to divide a table into two halves, maybe you really need two tables (and then see the “side-by-side” capabilities, Section <a href="section-side-by-side.ipynb" class="xref" alt="Section 22 Side-By-Side Panels" title="Section 22 Side-By-Side Panels">22</a>).</p></div>

<div class="mathbook-content"><p id="p-458">In the documentation for his excellent <span class="latex-logo">L<span class="A">a</span>T<span class="E">e</span>X</span> package, <a class="url" href="https://www.ctan.org/pkg/booktabs" target="_blank">booktabs</a>, Simon Fear gives two rules for what he calls “formal tables”: (1) Never, ever use vertical rules, and (2) Never use double rules.  We have resisted the temptation to enforce the former and have provided an alternative to the second (three thicknesses).  He refers to using tables for layout as creating “tableau.”  If you are finicky about the look of your work, the first three pages of the documentation is recommended reading.</p></div>

<div class="mathbook-content"><p id="p-467">Table cells with a fixed width where text wraps are known as “paragraph cells”. A cell will be created as a paragraph cell if and only if it has <code class="code-inline tex2jax_ignore"><p></code> children. And such cells should <em class="emphasis">only</em> have <code class="code-inline tex2jax_ignore"><p></code> children. The width of a paragraph cell is determined by a <code class="code-inline tex2jax_ignore">width</code> attribute on the corresponding <code class="code-inline tex2jax_ignore"><col></code> (as a percentage). If no <code class="code-inline tex2jax_ignore">width</code> is specified (or there isn't even a <code class="code-inline tex2jax_ignore"><col></code> in the first place) then <code class="code-inline tex2jax_ignore">xsltproc</code> will abort. If the column has a non-paragraph cell with contents that are wider than the paragraph cells, results will be undesirable. There is presently no implementation for a paragraph cell that has a <code class="code-inline tex2jax_ignore">colspan</code> greater than $1\text{,}$ although cells with <code class="code-inline tex2jax_ignore">colspan</code> greater than $1$ that are above or below a paragraph cell will behave. Setting <code class="code-inline tex2jax_ignore">width</code> on a <code class="code-inline tex2jax_ignore"><col></code> that has no paragraph cells may produce unexpected results. A <code class="code-inline tex2jax_ignore">valign</code> for the parent <code class="code-inline tex2jax_ignore"><row></code> (or the ambient <code class="code-inline tex2jax_ignore"><tabular></code>) can control vertical alignment (top, middle, or bottom). A paragraph cell's <code class="code-inline tex2jax_ignore">halign</code> attribute (left, center, right, or justify) controls how the text is justfied. Cells inherit <code class="code-inline tex2jax_ignore">halign</code> from <code class="code-inline tex2jax_ignore"><row></code>, <code class="code-inline tex2jax_ignore"><col></code>, and <code class="code-inline tex2jax_ignore"><tabular></code> in that order of preference. In a non-paragraph cell where <code class="code-inline tex2jax_ignore">halign='justify'</code>, the horizontal alignment will match the behavior of <code class="code-inline tex2jax_ignore">halign='left'</code>.</p></div>

<div class="mathbook-content"><p id="p-475">Table cells can have multiline content using <code class="code-inline tex2jax_ignore"><line></code> elements. This is not the same thing as a paragraph cell—line breaking will happen precisely where the author tells it to. A <code class="code-inline tex2jax_ignore"><line></code> will not break, even on a narrow screen. If a cell uses a <code class="code-inline tex2jax_ignore"><line></code>, it must only use a sequence of <code class="code-inline tex2jax_ignore"><line></code>s and no other content. As with paragraph cells, you can use a <code class="code-inline tex2jax_ignore">valign</code> attribute for the row.</p></div>
