<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 21 Units of Measure"><span class="type">Section</span><span class="codenumber">21</span><span class="title">Units of Measure</span></h2></div>

<div class="mathbook-content"><p id="p-687">Units of measure can be given xml treatment too with the <code class="code-inline tex2jax_ignore">quantity</code> element. In <span class="latex-logo">L<span class="A">a</span>T<span class="E">e</span>X</span>, the <code class="code-inline tex2jax_ignore">siunitx</code> package is loaded to achive unit handling. Since that package only offers SI units, some other common units will be added by MBX in the preamble. In HTML, the capabilities of <code class="code-inline tex2jax_ignore">siunitx</code> are simulated, weakly. Note that at present, you should not attempt to use the <code class="code-inline tex2jax_ignore">quantity</code> element within a math environment.</p></div>

<div class="mathbook-content"><p id="p-688">The value of gravitational constant $g$ is 9.8 <sup>m</sup>⁄<sub>s<sup>2</sup></sub>. Force is measured in <sup>kg m</sup>⁄<sub>s<sup>2</sup></sub>, also known as one N. A quantity with rather ridiculous units is 23 <sup>µha<sup>23</sup></sup>⁄<sub>°C s<sup>2</sup></sub>. One Hz is the same as <sup>1</sup>⁄<sub>s</sub>. You can have a unitless quantity, like 42, which may help with consistency between such numbers and units in the <span class="latex-logo">L<span class="A">a</span>T<span class="E">e</span>X</span> output. Some non-SI units are available, such as the absurd <sup>°F ft lb</sup>⁄<sub>gal</sub>.  The <span class="latex-logo">L<span class="A">a</span>T<span class="E">e</span>X</span> command <code class="code-inline tex2jax_ignore">\pi</code> is recognized within <code class="code-inline tex2jax_ignore">mag</code> in conversions to HTML, which is consistent with the behavior with a conversion to <span class="latex-logo">L<span class="A">a</span>T<span class="E">e</span>X</span>, for example there are 2$\pi$ rad in a full circle.</p></div>

<div class="mathbook-content"><p id="p-689">For a full list of the allowed units and prefixes, see <code class="code-inline tex2jax_ignore">mathbook-units.xsl</code>. If you have a need for more units, they need to be added to <code class="code-inline tex2jax_ignore">mathbook-units.xsl</code> in the section that deals with units which are not part of <code class="code-inline tex2jax_ignore">siunitx</code> by default.  Note that the <code class="code-inline tex2jax_ignore">mag</code> element should come first, followed by the <code class="code-inline tex2jax_ignore">unit</code> element, followed by the <code class="code-inline tex2jax_ignore">per</code> element.</p></div>
