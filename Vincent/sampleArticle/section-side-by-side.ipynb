<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 22 Side-By-Side Panels"><span class="type">Section</span><span class="codenumber">22</span><span class="title">Side-By-Side Panels</span></h2><a href="section-side-by-side.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"><h6 class="heading"><span class="title">Introduction</span></h6></div>

<div class="mathbook-content"><p id="p-690">The flow of a page is almost universally top-to-bottom.  But at times you would like to go <em class="emphasis">across</em> a page, perhaps to compare items (identical content in two different languages), or to make good use of a page real estate by grouping several small items together (e.g. images).  So the <code class="code-inline tex2jax_ignore"><sidebyside></code> tag is strictly a layout device, though it does convey some meaning by grouping certain objects together. A variety of different objects can be put side-by-side using the <code class="code-inline tex2jax_ignore">sidebyside</code> element.  Specifically, <code class="code-inline tex2jax_ignore">figure</code>, <code class="code-inline tex2jax_ignore">table</code>, <code class="code-inline tex2jax_ignore">listing</code>, <code class="code-inline tex2jax_ignore">paragraphs</code>, <code class="code-inline tex2jax_ignore">image</code>, <code class="code-inline tex2jax_ignore">tabular</code>, <code class="code-inline tex2jax_ignore">p</code>, <code class="code-inline tex2jax_ignore">ol</code>, <code class="code-inline tex2jax_ignore">ul</code>, <code class="code-inline tex2jax_ignore">dl</code>, <code class="code-inline tex2jax_ignore">pre</code>, <code class="code-inline tex2jax_ignore">poem</code>, and more.  The individual components of a <code class="code-inline tex2jax_ignore"><sidebyside></code> are generically called <dfn class="terminology">panels</dfn>.</p></div>

<div class="mathbook-content"><p id="p-691">As a layout device, the <code class="code-inline tex2jax_ignore"><sidebyside></code> does not allow a <code class="code-inline tex2jax_ignore"><caption></code>, is never numbered, and therefore cannot be cross-referenced.  You may cross-reference whatever element holds the <code class="code-inline tex2jax_ignore"><sidebyside></code>, and many of the panels may be cross-referenced individually.</p></div>

<div class="mathbook-content"><p id="p-692">As a first example, we have two single paragraphs, laid out with different widths, and slight margins on each side.  The widths have been chosen experimentally to get roughly identical heights for the two paragraphs of varying length.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection 22.1 Figures with Numbers Side-By-Side"><span class="type">Subsection</span><span class="codenumber">22.1</span><span class="title">Figures with Numbers Side-By-Side</span></h3></div>

<div class="mathbook-content"><p id="p-695">Figures, or other captioned items such as tables or listings, can be placed side-by-side using the <code class="code-inline tex2jax_ignore">sidebyside</code> element.  The figures will be captioned and numbered as if they were part of the vertical flow of the document.  For example, see <a href="section-side-by-side.ipynb#regular-figure" class="xref" alt="Figure 22.4 " title="Figure 22.4 ">Figure 22.4</a> and <a href="section-side-by-side.ipynb#another-regular-figure" class="xref" alt="Figure 22.5 " title="Figure 22.5 ">Figure 22.5</a></p></div>

<div class="mathbook-content"><p id="p-696">However, if the <code class="code-inline tex2jax_ignore"><sidebyside></code> is placed inside another <code class="code-inline tex2jax_ignore"><figure></code>, then the outer figure gets an overall caption and a “regular” number, while the captions of the interior items will be labelled as (a), (b), (c), etc; for example, see the subfigures in <a href="section-side-by-side.ipynb#fig-sidebyside-global" class="xref" alt="Figure 22.1 " title="Figure 22.1 ">Figure 22.1</a>. You can also cross-reference the subfigures individually, for example: <a href="section-side-by-side.ipynb#fig-sidebyside-subfigure" class="xref" alt="Figure 22.1.(a) " title="Figure 22.1.(a) ">Figure 22.1.(a)</a>.</p></div>

<div class="mathbook-content"><p id="p-697">The <code class="code-inline tex2jax_ignore">sidebyside</code> tag can have an attribute <code class="code-inline tex2jax_ignore">widths</code> that specifies a percentage width of the page for each panel of the layout.  There are automatic margins by default, and any remaining width is divided evenly to space out the panels.  When the <code class="code-inline tex2jax_ignore">margins</code> attribute is given as <code class="code-inline tex2jax_ignore">auto</code>, or in the default case, the margins provided each equal half of the inter-panel space.</p></div>

<div class="mathbook-content"><p id="p-698">With no attributes on the <code class="code-inline tex2jax_ignore">sidebyside</code>, each panel is the same width and there is no inter-panel space and no margin.  For a <code class="code-inline tex2jax_ignore">sidebyside</code> with a single panel, with its width specified, the panel will be centered.</p></div>

<div class="mathbook-content"><p id="p-710">Text can be put next to other blocks of text using either the <code class="code-inline tex2jax_ignore">paragraphs</code> element, which can contain multiple paragraphs using the <code class="code-inline tex2jax_ignore">p</code> element.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection 22.15 Side-By-Side Groups"><span class="type">Subsection</span><span class="codenumber">22.15</span><span class="title">Side-By-Side Groups</span></h3><a href="section-side-by-side.ipynb#subsection-sbsgroup" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-764">A “side-by-side group,” <code class="code-inline tex2jax_ignore"><sbsgroup></code>, is still in development.  (Notably, subcaptions do not behave as expected.)  It is a sequence of <code class="code-inline tex2jax_ignore">sidebyside</code>, which may conceivably use the same margins, widths and vertical alignments for each horizontal run of panels.  Attributes on the <code class="code-inline tex2jax_ignore">sbsgroup</code> are global to the group's enclosed <code class="code-inline tex2jax_ignore">sidebyside</code>, and will be used by each contained <code class="code-inline tex2jax_ignore">sidebyside</code>.  If attributes are present on an individual <code class="code-inline tex2jax_ignore">sidebyside</code>, they override the global values.  The next two examples demonstrate some of this behavior, in a limited way.</p></div>

<div class="mathbook-content"><p id="p-776">The main rationale for <code class="code-inline tex2jax_ignore">sbsgroup</code> is to layout a grid of items, and by placing the layout parameters on the <code class="code-inline tex2jax_ignore">sbsgroup</code> element, the items can line up across <code class="code-inline tex2jax_ignore">sidebyside</code> and subcaptioning can run across the whole group.  So, for example, if you have images to compare by placing in a grid, then making them all the same size, or of the same aspect ratio, can help with the overall consistency.</p></div>

<div class="mathbook-content"><p id="p-777">This example has three <code class="code-inline tex2jax_ignore">sidebyside</code>, each with four <code class="code-inline tex2jax_ignore">figure</code> containing an identical <code class="code-inline tex2jax_ignore">image</code>.  Since the images are identical and the <code class="code-inline tex2jax_ignore">width</code> is set to 20% they should all line up nicely with little effort.  Since the default for margins is automatic, the remaining 20% of the overall width will be used for three inter-panel spaces of 5% and two margins of 2.5% each.  Note the numbering of these as independent figures.  We have left the captions empty for reasons of space, but you could add more information.  Note that in print, a page break is allowed between any two of the <code class="code-inline tex2jax_ignore">sidebyside</code> and cannot be suppressed.</p></div>
