<div class="mathbook-content"><p id="p-637">Supporting a multitude of possible characters, across many languages and across many output formats can be a challenge.  One of our goals is to make this easier for authors.  Fortunately, the Unicode standard has led to improvements from the 7-bit ASCII standard of old.</p></div>

<div class="mathbook-content"><h6 class="heading"><span class="title">Unicode Characters for HTML Output</span></h6></div>

<div class="mathbook-content"><p id="p-638">First, we discuss HTML output.  If you include Unicode characters in your PreTeXt source, they should survive just fine <i class="foreign" lang="fr-FR">en route</i> to a web browser or e-reader.  Here are the caveats for HTML output:</p><ul class="disc"><li id="li-246"><p id="p-639">So that you can continue to get the best results with print and PDF output, use available empty elements for special characters, even if targeting HTML output, before resorting to a Unicode character.  For example, use <code class="code-inline tex2jax_ignore"><times /></code> for a small “multiplication sign” in text before resorting to the Unicode character <code class="code-inline tex2jax_ignore">U+00D7</code>.</p></li><li id="li-247"><p id="p-640">How you actually enter Unicode characters into your source file is dependent on your editor and operating system, and is therefore outside the scope of our documentation.  You can cut/paste characters and text from the source of our examples for initial testing and experimentation.</p></li><li id="li-248"><p id="p-641">Always, always identify your source as having Unicode characters by including the incantation <code class="code-inline tex2jax_ignore"><?xml version="1.0" encoding="UTF-8" ?></code> as the first line of your source file. (You <em class="emphasis">may</em> be able to accurately cut/paste this text here.  But if the copy has non-standard characters in it, go back to the top of this source file for a copy.)</p></li><li id="li-249"><p id="p-642">Alan Wood’s <a class="url" href="http://www.alanwood.net/unicode/unicode_samples.html" target="_blank">Unicode Resources</a> has a plethora of samples of various groups of Unicode characters.  If you, or your readers, are “missing” characters in a web browser, this is a good place to start testing the local setup.</p></li></ul></div>

<div class="mathbook-content"><h6 class="heading"><span class="title">Characters in <span class="latex-logo">L<span class="A">a</span>T<span class="E">e</span>X</span>, PDF, print</span></h6></div>

<div class="mathbook-content"><p id="p-643">The situation for <span class="latex-logo">L<span class="A">a</span>T<span class="E">e</span>X</span> is much more complicated, since <span class="latex-logo">T<span class="E">e</span>X</span> pre-dates Unicode's widespread adoption.</p></div>

<div class="mathbook-content"><p id="p-644">This sample article is intended to work well, out-of-the-box, for authors just starting with PreTeXt.  So we only include here examples that we know are likely to convert to PDF without any errors.  For more extensive examples and experiments, we provide the sample document <code class="code-inline tex2jax_ignore">examples/fonts/fonts-and-characters.xml</code>, so be aware of that example as you look to see what is possible.</p></div>

<div class="mathbook-content"><p id="p-645">Similarly, you should be able to process this sample article successfully with various <span class="latex-logo">L<span class="A">a</span>T<span class="E">e</span>X</span> engines. In principle, you should be able to use <code class="code-inline tex2jax_ignore">latex</code> (to produce a DVI), and possibly other (unsupported) engines, such as <code class="code-inline tex2jax_ignore">lualatex</code>.</p></div>

<div class="mathbook-content"><p id="p-646">Once you get beyond the Latin alphabet, with accents common in Western Europe and the Western Hemisphere, you will almost assuredly need to restrict your attention to producing PDF output with the <code class="code-inline tex2jax_ignore">xelatex</code> engine. <div class="mathbook-content"><p id="p-650">Note that the single and double quotes are upright and dumb, not curly and smart: <code class="code-inline tex2jax_ignore">' " ' " ' "</code>.  And a backtick is a backtick: <code class="code-inline tex2jax_ignore">` ` `</code>.  The zero is distinguished from the capital “oh”: <code class="code-inline tex2jax_ignore">0 O 0 O 0 O</code>.  And the numeral one is slightly different from the lower-case “ell”: <code class="code-inline tex2jax_ignore">1 l 1 l 1 l</code>.  The hyphen should be short and not expanded into some other kind of dash: <code class="code-inline tex2jax_ignore">- - -</code>.  These characters should all cut/paste out of a PDF into a text editor with no conversion to other characters.</p></div>
